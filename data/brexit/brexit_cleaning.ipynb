```python
import geopandas, topojson, matplotlib.pyplot as plt
```

```python
brexit = geopandas.read_file('./Local_Authority_Districts_December_2016_'
                             'Generalised_Clipped_Boundaries_in_the_UK_WGS84/'
                             'Local_Authority_Districts_December_2016_'
                             'Generalised_Clipped_Boundaries_in_the_UK_WGS84.shp')
```

```python
topology = topojson.Topology(brexit)
```

```python
simple_uk = topology.toposimplify(.025)
```

```python
f,ax = plt.subplots(1,1,figsize=(10,10))
simple_uk.to_gdf().plot(ax=ax)
```

```python
simple_uk.to_gdf().to_file('./simple_brexit.shp')
```

```python
f,ax = plt.subplots(1,1,figsize=(10,10))
topology.topoquantize(200).to_gdf().plot(ax=ax)
```
